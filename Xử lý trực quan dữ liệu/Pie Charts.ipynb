```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([10, 30, 40, 10, 10])
mylabels = ["Táo", "Xoài", "Lê", "Mận", "Ổi"]
plt.pie(y, labels = mylabels, startangle = 90) # tùy chỉnh độ quay
plt.show()
```

```python
# tach ra
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "cyan"] # có thể tùy chỉnh màu


plt.pie(y, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)# the duong
# thêm chú thích
plt.legend(title = "4 Fruits")
plt.show() 
```
