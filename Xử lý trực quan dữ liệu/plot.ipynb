```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([0, 6])
y_points = np.array([0, 255])
# draw a line a digram from (0,0)--(6,250)
plt.plot(x_points, y_points)
# draw 2 points 
plt.plot(x_points, y_points, 'o')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
a = np.array([0,1, 2, 3, 4, 5])
#Draw a line in a diagram from position (1, 3) to (2, 8) 
#then to (6, 1) and finally to position (8, 10):
plt.plot(x, y, linewidth ='10', color = 'r')
plt.plot(x, y, marker = 'o', ms = 10, mec = 'c', mfc = 'r')
# draw nét đứt
b = np.array([6,7,8,9,10])
plt.plot(b, '--')

plt.plot(a)
plt.plot(a, marker = '.')

# create grid
plt.grid() # có thể chỉ show dòng với trục x or y thêm () axis = 'x'
# thêm màu nét đứt thoải mái



# rename for label
plt.xlabel("trục y")
plt.ylabel("trục x")
plt.title("tiêu đề")
plt.show
```

```python
# display multiple plots: có thể chỉnh vị trí của 2 plots này top, under,...
# hoặc có thể vẽ đc nhiều plots tương tự có thể đặt tên title, axis x, y cho mỗi plot
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y, color = 'g')
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
# multiple color
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20, 50, 100, 200, 500, 1000, 60, 90, 10, 300, 600, 800, 15])
plt.scatter(x, y, c=colors, cmap='viridis')# cmap có thể sửa để tạo màu muốn có
plt.scatter(x, y, s = sizes, alpha = 0.5) # alpha là chỉnh độ mờ

plt.colorbar()

plt.show()
```

```python
# vẽ 1 bảng đa màu
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
