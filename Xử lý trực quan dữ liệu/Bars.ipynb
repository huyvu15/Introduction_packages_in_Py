```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([1, 2, 4, 5])

plt.bar(x, y, color = 'cyan', width = 0.3) # ngược lại thì dùng barh
plt.show()
```
